```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df= pd.read_csv('titanic.csv')
df
```

```python
df.info()
```

```python
import matplotlib as plt
import seaborn as sns
%matplotlib inline
sns.factorplot(x ='Sex', kind = 'count', data = df,hue='Pclass')

```

```python
sns.factorplot(x ='Pclass', kind = 'count', data = df,hue='Sex')
```

```python
def male_female_child(passenger):
    age ,sex=passenger
    if (age<16):
       return 'child'
    else:
         if(sex=='male'):
                return 'man'
         else:
                return 'women'
```

```python
df['person']=df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
sns.factorplot(x ='Pclass', kind = 'count', data = df,hue='person')
```

```python
fig = sns.FacetGrid(df,hue='Sex',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest=df['Age'].max()
fig.set(xlim=(0,oldest))
        
fig.add_legend()
```

```python
fig = sns.FacetGrid(df,hue='person',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest=df['Age'].max()
fig.set(xlim=(0,oldest))
        
fig.add_legend()
```

```python
 fig = sns.FacetGrid(df,hue='Pclass',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest=df['Age'].max()
fig.set(xlim=(0,oldest))
        
fig.add_legend()
```

```python
deck = df['Cabin'].dropna()
deck.head()
```
